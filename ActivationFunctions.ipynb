```python
# Sigmoid Activation Function
# Formula : sigmoid(x) = 1 / 1 + e^-x

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

input_x = np.array([-10, -1, 0, 1, 10])
output_of_x = sigmoid(input_x)

print("Sigmoid Input:", input_x)
print("Sigmoid Output:", output_of_x)
```

```python
# Softmax Activation Function
# Formula: softmax(x_i) = exp(x_i) / sum(exp(x_j)) for all j

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

input_x = np.array([1.0, 2.0, 3.0])
output_of_x = softmax(input_x)

print("Softmax Input:", input_x)
print("Softmax Output:", output_of_x)
```

```python
# ReLU Activation Function
# Formula: ReLU(x) = max(0, x)

def relu(x):
    return np.maximum(0, x)

input_x = np.array([-10, -1, 0, 1, 10])
output_of_x = relu(input_x)

print("ReLU Input:", input_x)
print("ReLU Output:", output_of_x)
```

```python
# Leaky ReLU Activation Function
# Formula: LeakyReLU(x) = x if x > 0 else alpha * x

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

input_x = np.array([-10, -1, 0, 1, 10])
output_of_x = leaky_relu(input_x)

print("Leaky ReLU Input:", input_x)
print("Leaky ReLU Output:", output_of_x)
```

```python
# Tanh Activation Function
# Formula: tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))

def tanh(x):
    return np.tanh(x)

input_x = np.array([-10, -1, 0, 1, 10])
output_of_x = tanh(input_x)

print("Tanh Input:", input_x)
print("Tanh Output:", output_of_x)
```
