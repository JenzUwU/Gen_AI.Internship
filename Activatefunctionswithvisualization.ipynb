```python
# Sigmoid Activation Function
# Formula: sigmoid(x) = 1 / (1 + e^-x)

import numpy as np
import matplotlib.pyplot as plt   # MATLAB-style plotting

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

input_x = np.linspace(-10, 10, 100)
output_y = sigmoid(input_x)

plt.figure()
plt.plot(input_x, output_y, 'b-', linewidth=2)
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
# ReLU Activation Function
# Formula: ReLU(x) = max(0, x)

def relu(x):
    return np.maximum(0, x)

input_x = np.linspace(-10, 10, 100)
output_y = relu(input_x)

plt.figure()
plt.plot(input_x, output_y, 'g-', linewidth=2)
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
# Leaky ReLU Activation Function
# Formula: LeakyReLU(x) = x if x > 0 else alpha * x

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

input_x = np.linspace(-10, 10, 100)
output_y = leaky_relu(input_x, alpha=0.1)

plt.figure()
plt.plot(input_x, output_y, 'c-', linewidth=2)
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
# Tanh Activation Function
# Formula: tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))

def tanh(x):
    return np.tanh(x)

input_x = np.linspace(-10, 10, 100)
output_y = tanh(input_x)

plt.figure()
plt.plot(input_x, output_y, 'm-', linewidth=2)
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
# Softmax Activation Function
# Formula: softmax(x_i) = exp(x_i) / sum(exp(x_j)) for all j

def softmax(x):
    e_x = np.exp(x - np.max(x))   # stability trick
    return e_x / e_x.sum()

input_x = np.array([1.0, 2.0, 3.0, 4.0])
output_y = softmax(input_x)

plt.figure()
plt.bar(range(len(input_x)), output_y, color='r')
plt.title('Softmax Activation Function')
plt.xlabel('Class Index')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```
